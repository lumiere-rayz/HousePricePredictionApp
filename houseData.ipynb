```python
import numpy as np 
import pandas as pd

houseData = pd.read_excel(io.BytesIO(uploaded['house_data.xlsx']))
houseData.head()
```

```python
houseData.describe() # descriptive analyses
```

```python
houseData.dtypes #print column data types
```

```python
print(houseData.shape[0])#print number of rows
print(houseData.shape[1])#print number of columns
```

```python
houseData = houseData.drop(['id', 'zipcode', 'date', 'lat','long'], axis=1)
```

```python
houseData.columns
```

```python
houseData.corr()
```

```python
houseData = houseData.drop(['sqft_above'], axis=1)
```

```python
houseData.corr()
```

```python
houseData.head()
```

```python
houseData = houseData.drop(['yr_renovated'], axis=1)
```

```python
bins = [1900, 1915 , 1930, 1945, 1960, 1975, 1990, 2005, 2020] #binning
labels = [1,2,3,4,5,6,7,8]
houseData['yr_built_binned'] = pd.cut(houseData['yr_built'], bins=bins, labels=labels)
```

```python
houseData.head()
```

```python
houseData = houseData.drop(['yr_built'], axis=1)
```

```python
houseData.head()
```

```python
houseData = houseData.drop(['waterfront'], axis=1)
```

```python
cat = ['yr_built_binned'] #one-hot encoding

for var in cat:
    houseData = pd.concat([houseData, pd.get_dummies(houseData[var], prefix=var)], axis=1)
    del houseData[var]
```

```python
houseData.head()
```

```python
data_columns = houseData.columns #variable selection
X = houseData[data_columns[data_columns != 'price']]
y = houseData['price']
```

```python
print(X)
```

```python
print(y)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split #spliting data
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y,test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor #RandomForest Machine Learning model
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error

forest = RandomForestRegressor()
pipeReg = Pipeline([('Scaler', StandardScaler()), ('regressor', RandomForestRegressor())])
pipeReg.fit(X_train, y_train)

# making predictions
yPredF = pipeReg.predict(X_test)
mean_squared_error(y_test, yPredF)

print(pipeReg.score(X_test, y_test))
```

```python
from sklearn import linear_model 
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR

# Import metrics systems
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.metrics import r2_score
```

```python
linear = linear_model.LinearRegression() #linear model
tree = DecisionTreeRegressor()
svr = SVR(kernel='rbf',C=8)

linear.fit(X_train, y_train)
tree.fit(X_train, y_train)
svr.fit(X_train, y_train)

linear_predict = pd.DataFrame(linear.predict(X_test))
tree_predict = pd.DataFrame(tree.predict(X_test))
svr_predict = pd.DataFrame(svr.predict(X_test))
y_real = pd.DataFrame(y_test)


new_df = pd.DataFrame() 

new_df[['y_real']] = y_real
new_df[['linear_predict']] = linear_predict
new_df[['tree_predict']] = tree_predict
```

```python
new_df[['svr_predict']] = svr_predict

print(new_df.head())

print('linear score:{0}'.format(linear.score(X_test, y_test)))
print('tree score:{0}'.format(tree.score(X_test, y_test)))
print('svr score:{0}'.format(svr.score(X_test, y_test)))

err_lin = mean_squared_error(y_real, linear.predict(X_test))
err_tree = mean_squared_error(y_real, tree.predict(X_test))
err_svr = mean_squared_error(y_real, svr.predict(X_test))

print(err_lin, err_tree, err_svr)
```
